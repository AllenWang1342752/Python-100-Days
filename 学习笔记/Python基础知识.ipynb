### 面向对象
#### 面向对象的含义  
面向对象编程是一种编程范式（即程序设计的方法论）。
在面向对象编程的世界里，程序中的**数据和操作数据的函数是一个逻辑上的整体**，称之为**对象**。  

对象可以接受消息，解决问题的方法就是创建对象并向对象发出各种各样的消息；通过消息传递，程序中的多个对象可以协同工作，这样就能构造出复杂的系统并解决现实中的问题。
> 面向对象编程：把一组数据和处理数据的方法组成**对象**，把行为相同的对象归纳为**类**，通过**封装**隐藏对象的内部细节，通过**继承**实现类的特化和泛化，通过**多态**实现基于对象类型的动态分派。
> - 封装：隐藏内部实现细节，往外暴露简单调用接口。例如在类中定义对象的方法就是封装，给对象发一个消息（即不用明确具体实现）就可以执行方法中的代码。
#### 面向对象编程三步走
第一步：定义类  
第二步：创建对象  
第三步：给对象发消息
Python内置的list、set、dict都是提前定义好的类，可以跳过第一步；  
如果第一步、第二部都提前走了，就是某些内置对象，可以“开箱即用”。

#### 类和对象
含义：类是对象的蓝图和模板(例如人类)，对象是类的实例（例如具体的人），是可以接受消息的实体。   
> 一切皆为对象，对象都有属性和行为，每个对象都是独一无二的，而且对象一定属于某个类。  
> 接收消息是什么意思？

属性：属性是对象的静态特征，行为（方法）是对象的动态特征。具有共同特征的对象和属性和行为抽取出来就可以定义一个类。

类的定义和使用：

```python
### 类的创建 ###
class Student:

    def __init__(self,name,age): # 初始化方法,即常见对象时，
        #首先会在内存中获取对象的所需的内存空间，然后会自动执行该方法，完成对内存的初始化操作
        self.name = name # 对象属性赋初始化值
        self.age = age # = 右侧的age是传入的参数

    def study(self,course_name):
        print(f'{self.name}正在学习{course_name}.')

    def play(self):
        print(f'{self.name}学生正在玩游戏.')

### 类的使用 ### 
stu1 = Student('王克章',28) # 构造器语法创建类的对象,此时会传入类的初始化参数
stu2 = Student('王大锤',25)
print(stu1) 
# 输出 <__main__.Student object at 0x000002459692B2C0>
# 上面这个输出代表是一个Student类的对象，以及这个对象的以十六进制表示的内存地址
# __main__表示Student类定义在主模块中（当前脚本）
print(stu2) 
# 输出 <__main__.Student object at 0x000002459692B2C0>
print(hex(id(stu1)),hex(id(stu2)))
# 函数hex()将整数转为十六进制字符串
# 函数id()返回对象在内存中的唯一标识(一般是内存地址)，该标识在对象的全生命周期内保持一致
stu3 = stu2 # 并没有创建新的对象
# 变量存的是一个对象在内存中的逻辑地址（位置）

### 调用对象的方法 ###
Student.study(stu1,'Python程序设计')# 类.方法(接收消息的对象,方法入参) 来调用
Student.play(stu2)
stu1.study('Python程序设计')# 对象.方法(仅为方法入参)来调用
stu2.play()
```

> 写在类里面的函数我们通常称之为方法，方法就是对象的行为，也就是对象可以接收的消息。方法的第一个参数通常都是self，它代表了接收这个消息的对象本身。
